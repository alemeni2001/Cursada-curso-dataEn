### 1.4
Dada la siguiente lista de ``mails_viejos``, pasarla de forma automatizada a la de ``mails_nuevos``

```python
# Tengo mi lista de correos viejos 
mails_viejos = ['Rihanna@hotmail.com', 'lady_gaga@hotmail.com', 'taylor_swift_red@hotmail.com']

# Creo mi lista donde voy a guardar los nuevos mails modificados
mails_nuevos = []

for correo in mails_viejos:
    #Con la funcion replace cambio la parte del @hotmail.com por @gmail.com
    correo_actualizado = correo.replace("@hotmail.com", "@gmail.com")
    
    #Lo agrego con el metodo append en la lista vacia creada
    mails_nuevos.append(correo_actualizado)

print(mails_nuevos)    
#mails_nuevos = ['Rihanna@gmail.com', 'lady_gaga@gmail.com', 'taylor_swift_red@gmail.com']
```

### 2.3

Escribir un programa que pida al usuario una palabra y muestre por pantalla el número de veces que contiene cada vocal.


```python
palabra = input("Ingrese una palabra: ")
palabra = palabra.lower()

contador_de_vocales = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0}

for letra in palabra:
    if letra in contador_de_vocales:
        contador_de_vocales[letra] += 1

for vocal, cantidad in contador_de_vocales.items():
    print(f"La vocal '{vocal}' aparece {cantidad} veces.")
```

### 2.5

Escribir un programa que pregunte una fecha en formato dd/mm/aaaa y muestre por pantalla la misma fecha en formato dd de <mes> de aaaa donde <mes> es el nombre del mes.

```python
meses = {
    1: "enero",
    2: "febrero",
    3: "marzo",
    4: "abril",
    5: "mayo",
    6: "junio",
    7: "julio",
    8: "agosto",
    9: "septiembre",
    10: "octubre",
    11: "noviembre",
    12: "diciembre"
}

fecha_ingresada = input("Ingrese una fecha en formato dd/mm/aaaa: ")

if "/" in fecha_ingresada:
    dia, mes, anio = fecha_ingresada.split("/")
    if dia.isdigit() and mes.isdigit() and anio.isdigit():
        nombre_mes = meses.get(int(mes))
        if nombre_mes:
            print(f"{dia} de {nombre_mes} de {anio}")
        else:
            print("Error: Mes inválido.")
    else:
        print("Error: Ingrese una fecha válida en formato dd/mm/aaaa.")
else:
    print("Error: Ingrese una fecha válida en formato dd/mm/aaaa.")
```

Escribir una funcion que tome un string como parametro y devuelva la palabra mas larga en dicho string


```python
def palabra_mas_larga(string):
    palabras = string.split()
    palabra_mas_larga = ""

    for palabra in palabras:
        if len(palabra) > len(palabra_mas_larga):
            palabra_mas_larga = palabra

    return palabra_mas_larga

string = str("Hola como andas todo bien ")
palabra_mas_larga(string)
```

### 3.1
Escribir un programa que reciba una cadena de caracteres y devuelva un diccionario con cada palabra que contiene y su frecuencia.


```python
def contar_palabras(cadena):
    palabras = cadena.lower().split()
    frecuencias = {}

    for palabra in palabras:
        if palabra in frecuencias:
            frecuencias[palabra] += 1
        else:
            frecuencias[palabra] = 1

    return frecuencias

texto = "Este es un texto de ejemplo para contar palabras repetidas en un texto"
resultado = contar_palabras(texto)
print(resultado)
```

```python
# CODIGO CORREGIDO
listin = {'Juan': 123456789, 'Pedro': 987654321}

def elimina(listin, usuario):
    if usuario in listin:
        del listin[usuario]
    return listin

print(elimina(listin, 'Pedro'))
```

A simple vista se puede observar que se quiere eliminar el usuario Pablo, el cual no se encuentra dentro del diccionario. El codgio deberia eliminar a Juan o a Pedro que si pertenecen. Para resolver esto se decidio poner un condicional en caso de que el usuario no pertenezca dentro del diccionario. Finalmente la funcion deberia devolver el diccionario sin el usuario que se haya querido eliminar por eso retorno listin, en el otro caso estaria devolviendo el que quiero eliminar y al no encontrarse mas en el diccionario este dara un error.
