- ``embarked``: the persons place of of embarkment.
- ``country``: the persons home country.
- ``ticketno``: the persons ticket number (NA for crew members).
- ``fare``: the ticket price (NA for crew members, musicians and employees of the shipyard company).
- ``sibsp``: a number specifying the number if siblings/spouses aboard.
- ``parch``: a number specifying the number of parents/children aboard.
- ``survived``: a string ("no" or "yes") specifying whether the person has survived the sinking.

Quiero crear una funcion que me transforme de mi dataset los "yes" en True y los "no" en False. Guardaria una columna nueva y borraria la anterior ya que poner un booleano me parece mucho mas descriptivo que un string en el caso de que haya sobrevivido. Una vez realizado esto procedo a borrar mi  columna survived

````python
def transformar_supervivencia(valor):
    return True if valor == 'yes' else False
````

````python
cabinas = ["S","C","B","Q"]
titanic_cabinas = titanic.loc[titanic['embarked'].isin(cabinas),:]

titanic_cabinas['supervivencia_bool'] = titanic_cabinas['survived'].apply(transformar_supervivencia)

# Eliminar la columna 'survived'
titanic_cabinas.drop(columns=['survived'], inplace=True)

titanic_cabinas.groupby('embarked')['supervivencia_bool'].mean().round(2)

````
